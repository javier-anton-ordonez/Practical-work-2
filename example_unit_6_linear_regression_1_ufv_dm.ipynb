```python
import numpy as np
```

```python
# Dataset
x = np.array([2, 3, 5, 7, 8])  # Ages
y = np.array([14, 20, 32, 42, 44])  # Weights
```

```python
# Calculating the means of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculating b1
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)
b1 = numerator / denominator

# Calculating b0
b0 = y_mean - b1 * x_mean
```

```python
print(f"b1 (slope of the regression line): {b1}")
print(f"b0 (y-intercept of the regression line): {b0}")
```
